```python
tumor_types = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
train_path = 'data/Training/'
```

```python
%%time
X_train = []
y_train = []
for t_ind, tumor in enumerate(tumor_types):
    for f_ind, filename in enumerate(os.listdir(train_path+tumor+'/')):

        image = Image.open(train_path+tumor+'/'+filename)
        X_train.append(np.array(image))
        y_train.append(t_ind)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
for i in range(X_train.shape[0]):
    if X_train[i].shape != (512, 512, 3):
        print(i, X_train[i].shape, y_train[i])
```

```python
for i in range(X_train.shape[0]):
    if X_train[i].shape == (512, 512, 3) and y_train[i] == 3:
        print(i, X_train[i].shape, y_train[i])
```

```python
X_train_indices = [i for i in range(len(X_train)) if X_train[i].shape == (512, 512, 3)]
X_tr_std = np.array(X_train[X_train_indices])
y_tr_std = y_train[X_train_indices]
```

```python
X_tr_std = np.array(list(X_tr_std))
```

```python
X_tr_std.shape
```

```python
for i in range(X_tr_std.shape[0]):
    if X_tr_std[i].shape != (512, 512, 3):
        print(i, X_tr_std[i].shape, y_tr_std[i])
```

```python
X_tr_std_flat = X_tr_std.reshape(X_tr_std.shape[0], X_tr_std.shape[1] * X_tr_std.shape[2] * X_tr_std.shape[3])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=10)
```

```python
rf.fit(X_tr_std_flat, y_tr_std)
```

```python
rf.score(X_tr_std_flat, y_tr_std)
```
